```python
from timeit import default_timer as timer

#F!
def F1(val):
  if (val <= 2):
    return 1
  else:
    return (F1(val-1)+F1(val-2))

#F2
def F2(val):
  if (val <= 2):
    return 1
  else:
    p = 1
    q = 1
    r = 0
    for i in range(3, val+1):
      r = p + q
      p = q
      q = r
    return r  

#F3
def powerofG (n):
  G = [[0, 1],[1, 1]]
  if(n == 1):
    return G
  else :
    if (n%2 == 0):
      H = powerofG(n/2)
      return (matrixMul(H, H))
    else:
      H = powerofG((n-1)/2)
      temp = (matrixMul(H, H))
      return (matrixMul(temp, G))

def matrixMul(X, Y):
  result = [[0,0],[0,0]]
  for i in range((len(X))):
    for j in range(len(Y[0])):
      for k in range(len(Y)):
        result[i][j] += X[i][k] * Y[k][j]
  return result


def F3(n):
    temp = powerofG(n)
    return (temp[0][1])
```

```python
Y1 = []
for i in datasetinput1:
    start = timer()
    v = F1(i)
    print(v)
    end = timer()
    Y1.append(end-start)
```

```python
Y2 = []
for i in datasetinput2:
    start = timer()
    v = F2(i)
    end = timer()
    Y2.append(end-start)
print(Y2)
```

```python
Y3 = []
for i in datasetinput2:
    start = timer()
    v = F3(i)
    end = timer()
    Y3.append(end-start)
print(Y3)
```

```python
#graph1
plt.scatter(datasetinput1, Y1, color='red')
plt.plot(datasetinput1, Y1, color='blue')
plt.title('GRAPH 1')
plt.xlabel('n')
plt.ylabel('Time')
plt.show
```
